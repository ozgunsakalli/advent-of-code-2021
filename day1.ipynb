Advent of Code 2021 - Day 1

Part 1: How many measurements are larger than the previous measurement?

```python
with open('input_day1.txt', 'rt') as f:
    data = f.readlines()

data = [int(item.rstrip()) for item in data]

print(data[:10])
print(data[-10:])
print(len(data))
```

```python
counter = 0
for n, meas in enumerate(data[1:]):
    print(meas, data[n], bool(meas > data[n]))
    if meas > data[n]:
        counter += 1
```

```python
# Part 1 answer
print(counter)
```

Part 2: Consider sums of a three-measurement sliding window. How many sums are larger than the previous sum?

```python
counter = 0
for n, _ in enumerate(data[3:]):
    prev = sum(data[n:n+3])
    curr = sum(data[n+1:n+4])
    if curr > prev:
        counter += 1
```

```python
# Part 2 answer
print(counter)
```
