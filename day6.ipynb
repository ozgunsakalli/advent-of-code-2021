Advent of Code 2021 - Day 6

Part 1:
Find a way to simulate lanternfish. How many lanternfish would there be after 80 days?


```python
from collections import Counter
from collections import defaultdict
```

```python
with open('input_day6.txt', 'rt') as f:
    data = f.read()

data = [int(item) for item in data.split(',')]
data[:10]
```

```python
class Lanternfish:

    def __init__(self, timer, population:list):
        self.timer = timer
        self.population = population
        self.population.append(self)

    def increment_day(self):
        if self.timer > 0:
            self.timer -= 1
        else:
            self.timer = 6
            Lanternfish(8, self.population)
```

```python
pop = []

for timer in data:
    Lanternfish(timer, pop)

for i in range(80):
    print('Day: ', i, '\tPopulation: ', len(pop))
    for fish in pop[:len(pop)]:
        fish.increment_day()
```

```python
# Part 1 answer
print(len(pop))
```

Part 2: How many lanternfish would there be after 256 days?

```python
class Population:

    def __init__(self, timers):
        init_pop = defaultdict(int)
        for key, value in Counter(timers).items():
            init_pop[key] += value
        self.timers = init_pop
    
    def increment_day(self):
        for key, value in self.timers.copy().items():
            if key > 0:
                self.timers[key] -= value
                self.timers[key-1] += value
            else:
                self.timers[8] += value
                self.timers[6] += value
                self.timers[0] -= value
            
```

```python
pop = Population(data)

for i in range(256):
    print('Day: ', i)
    # print(pop.timers)
    pop.increment_day()
```
